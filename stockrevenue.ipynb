```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
tsla = yf.Ticker("TSLA")
```

```python
#Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function. 
#Upload a screenshot of the results and code from the beginning of Question 1 to the results below.
tesla_data = tsla.history(period='max').reset_index()
tesla_data.head(n=5)
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data
#Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.
r = requests.get('https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue')
tesla_revenue = pd.read_html(r.text)[0]
tesla_revenue.tail(n=5)
```

```python
#Question 3: Use yfinance to Extract Stock Data
#Reset the index, save, and display the first five rows of the gme_data dataframe using the head function. 
#Upload a screenshot of the results and code from the beginning of Question 1 to the results below.

gme = yf.Ticker("GME")

gme_data = gme.history(period='max').reset_index()
gme_data.head(n=5)
```

```python
#Question 4: Use Webscraping to Extract GME Revenue Data
#Display the last five rows of the gme_revenue dataframe using the tail function. Upload a screenshot of the results
r = requests.get('https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue')
gme_revenue = pd.read_html(r.text)[0]
gme_revenue.tail(n=5)
```

```python
#Question 5: Plot Tesla Stock Graph
#Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.
graph_tesla = make_subplots(specs=[[{"secondary_y": True}]])
graph_tesla.add_trace(
    go.Candlestick(
        x=tesla_data['Date'],
        open = tesla_data['Open'],
        close = tesla_data['Close'],
        high = tesla_data['High'],
        low = tesla_data['Low'],
        name = 'Stock'
    ),
)

graph_tesla.show()
```

```python
#Question 6: Plot GameStop Stock Graph
#Use the make_graph function to graph the GameStop Stock Data, also provide a title for the graph.
graph_gamestop = make_subplots(specs=[[{"secondary_y": True}]])
graph_gamestop.add_trace(
    go.Candlestick(
        x=gme_data['Date'],
        open = gme_data['Open'],
        close = gme_data['Close'],
        high = gme_data['High'],
        low = gme_data['Low'],
        name = 'Stock'
    ),
)

graph_gamestop.show()
```
